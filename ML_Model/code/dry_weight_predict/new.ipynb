```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd

# Load data
df = pd.read_excel('../../data_set/hd_sessions/cleaned_session_data.xlsx')

# Create target and features
df['Next Dry weight (kg)'] = df.groupby('Subject_ID')['Dry weight (kg)'].shift(-1)
df['Dry weight adjustment (kg)'] = df['Next Dry weight (kg)'] - df['Dry weight (kg)']
df = df.dropna(subset=['Dry weight adjustment (kg)']).reset_index(drop=True)
# Ensure the dry weight adjustment is between -5 and 5 kg
df = df[(df['Dry weight adjustment (kg)'] >= -5) & (df['Dry weight adjustment (kg)'] <= 5)]

# Rule-based features
df['High_SBP'] = df['SYS (mmHg)'] > 140
df['UFR'] = df['PUF (ml)'] / (df['HD duration (h)'] * df['Pre HD weight (kg)'])
df['UFR_below_15'] = df['UFR'] < 15

df['High_SBP'] = df['High_SBP'].astype(int)
df['UFR_below_15'] = df['UFR_below_15'].astype(int)

# Binary adjustment class
def get_adjustment_class(row, threshold=0.1):
    diff = row['Next Dry weight (kg)'] - row['Dry weight (kg)']
    if abs(diff) >= threshold:
        return 1
    else:
        return 0

df['Adjustment_Class'] = df.apply(get_adjustment_class, axis=1)
df['Adjustment_Direction'] = df['Dry weight adjustment (kg)'].apply(
    lambda x: 1 if x >= 0.5 else (2 if x <= -0.5 else 0)
)

# Additional features
df['Weight_gain_pct'] = df['Weight gain (kg)'] / df['Dry weight (kg)'] * 100

df['Weight_gain_avg_3'] = df.groupby('Subject_ID')['Weight gain (kg)'].transform(
    lambda x: x.rolling(3, min_periods=1).mean()
)
df['SYS_avg_3'] = df.groupby('Subject_ID')['SYS (mmHg)'].transform(
    lambda x: x.rolling(3, min_periods=1).mean()
)


# Outlier handling
df = df[df['UFR'].between(0, 20)]
df = df[df['Weight gain (kg)'] >= 0]
df = df[df['SYS (mmHg)'].between(50, 250)].reset_index(drop=True)


# Selected features 
features = [
    'AP (mmHg)', 'AUF (ml)', 'High_SBP', 'BFR (ml/min)', 'SYS_avg_3', 'HD duration (h)', 'UFR',
    'PUF (ml)', 'TMP (mmHg)', 'VP (mmHg)', 'Weight gain (kg)', 'SYS (mmHg)', 'DIA (mmHg)',
    'Pre HD weight (kg)', 'Post HD weight (kg)', 'Dry weight (kg)', 'Weight_gain_avg_3', 'Weight_gain_pct'
]

# Prepare data
X = df[features].fillna(df[features].median())
y = df['Adjustment_Class']
original_indices = df.index

# Print class distribution
print("Class Distribution:\n", df['Adjustment_Class'].value_counts())


# Train a Random Forest model to get feature importance
rf = RandomForestClassifier(random_state=42)
rf.fit(X, y)

# Create a DataFrame of feature importance
feature_importance = pd.DataFrame({
    'Feature': features,
    'Importance': rf.feature_importances_
}).sort_values(by='Importance', ascending=False)

# Print feature importance
print("Feature Importance:\n", feature_importance)

# Select top features (e.g., top 15 or based on a threshold)
top_features = feature_importance['Feature'].head(15).tolist()  # Adjust number as needed
print("Selected Top Features:\n", top_features)

# Update X to include only top features
X_selected = X[top_features]
```

```python
import pandas as pd
import numpy as np
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.model_selection import KFold, GridSearchCV
from sklearn.metrics import classification_report, roc_auc_score
from imblearn.combine import SMOTEENN
import warnings
warnings.filterwarnings('ignore')

# Custom focal loss for XGBoost (scikit-learn compatible)
def focal_loss_xgb(alpha=0.75, gamma=2.0):
    def focal_loss_obj(y_true, y_pred):
        # y_true: NumPy array of true labels
        # y_pred: NumPy array of raw predictions (logits)
        p = 1.0 / (1.0 + np.exp(-y_pred))  # Sigmoid
        p_t = np.where(y_true == 1, p, 1 - p)
        fl = -alpha * (1 - p_t) ** gamma * np.log(p_t + 1e-10)
        grad = alpha * gamma * (1 - p_t) ** (gamma - 1) * p_t * np.log(p_t + 1e-10) * (y_true - p)
        hess = alpha * gamma * (1 - p_t) ** (gamma - 1) * (np.log(p_t + 1e-10) * (1 - p_t) + gamma * p * (1 - p))
        return grad, hess
    return focal_loss_obj

# Custom focal loss for LightGBM (scikit-learn compatible)
def focal_loss_lgb(y_true, y_pred, alpha=0.75, gamma=2.0):
    # y_true: NumPy array of true labels
    # y_pred: NumPy array of raw predictions (logits)
    p = 1.0 / (1.0 + np.exp(-y_pred))  # Sigmoid
    p_t = np.where(y_true == 1, p, 1 - p)
    grad = -alpha * gamma * (1 - p_t) ** (gamma - 1) * (y_true - p)
    hess = alpha * gamma * (1 - p_t) ** (gamma - 1) * (gamma * p * (1 - p) + (1 - p_t))
    return grad, hess

def focal_loss_lgb_eval(y_true, y_pred, alpha=0.75, gamma=2.0):
    # y_true: NumPy array of true labels
    # y_pred: NumPy array of raw predictions (logits)
    p = 1.0 / (1.0 + np.exp(-y_pred))
    p_t = np.where(y_true == 1, p, 1 - p)
    loss = -alpha * (1 - p_t) ** gamma * np.log(p_t + 1e-10)
    return 'focal_loss', np.mean(loss), False

# Assuming X, y, and top_15_features are defined from your code
top_15_features = [
    'SYS_avg_3', 'VP (mmHg)', 'Weight_gain_avg_3', 'Post HD weight (kg)', 
    'Pre HD weight (kg)', 'AP (mmHg)', 'Weight_gain_pct', 'SYS (mmHg)', 
    'UFR', 'TMP (mmHg)', 'DIA (mmHg)', 'Dry weight (kg)', 
    'AUF (ml)', 'Weight gain (kg)', 'PUF (ml)'
]
X_selected = X[top_15_features]

# Initialize SMOTEENN
smoteenn = SMOTEENN(random_state=42, sampling_strategy=1.0)  # Equalize class 0 and class 1

# Initialize base models with focal loss
xgb_model = XGBClassifier(
    objective=focal_loss_xgb(alpha=0.75, gamma=2.0),
    random_state=42,
    eval_metric='logloss'  # Placeholder, as focal loss is custom
)
lgb_model = LGBMClassifier(
    objective="binary",
    random_state=42
)

# Define hyperparameter grids
xgb_param_grid = {
    'max_depth': [3, 5],
    'learning_rate': [0.05, 0.1],
    'n_estimators': [100, 200]
}
lgb_param_grid = {
    'num_leaves': [15, 31],
    'max_depth': [5, 7],
    'learning_rate': [0.05, 0.1],
    'n_estimators': [100, 200]
}

# Cross-validation with standard KFold (no grouping)
kf = KFold(n_splits=5, shuffle=True, random_state=42)
for fold, (train_idx, val_idx) in enumerate(kf.split(X_selected, y)):
    X_train, X_val = X_selected.iloc[train_idx], X_selected.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]
    
    # Apply SMOTEENN to training data
    X_train_resampled, y_train_resampled = smoteenn.fit_resample(X_train, y_train)
    
    # Tune XGBoost
    xgb_grid = GridSearchCV(
        estimator=xgb_model,
        param_grid=xgb_param_grid,
        scoring='precision_macro',
        cv=3,
        n_jobs=-1,
        error_score='raise'
    )
    xgb_grid.fit(X_train_resampled, y_train_resampled)
    
    # Tune LightGBM
    lgb_grid = GridSearchCV(
        estimator=lgb_model,
        param_grid=lgb_param_grid,
        scoring='precision_macro',
        cv=3,
        n_jobs=-1,
        error_score='raise'
    )
    lgb_grid.fit(X_train_resampled, y_train_resampled)
    
    # Create ensemble
    ensemble_model = VotingClassifier(
        estimators=[
            ('xgb', xgb_grid.best_estimator_),
            ('lgb', lgb_grid.best_estimator_)
        ],
        voting='soft'
    )
    
    # Train ensemble
    ensemble_model.fit(X_train_resampled, y_train_resampled)
    
    # Predict and evaluate
    y_pred = ensemble_model.predict(X_val)
    y_scores = ensemble_model.predict_proba(X_val)[:, 1]
    
    print(f"\nFold {fold + 1} Results:")
    print(f"XGBoost Best Parameters: {xgb_grid.best_params_}")
    print(f"LightGBM Best Parameters: {lgb_grid.best_params_}")
    print(classification_report(y_val, y_pred))
    print(f"ROC-AUC: {roc_auc_score(y_val, y_scores):.3f}")

# Train final ensemble model
X_resampled, y_resampled = smoteenn.fit_resample(X_selected, y)
final_model = VotingClassifier(
    estimators=[
        ('xgb', xgb_grid.best_estimator_),
        ('lgb', lgb_grid.best_estimator_)
    ],
    voting='soft'
)
final_model.fit(X_resampled, y_resampled)

# Save feature importance (average of XGBoost and LightGBM)
xgb_importance = xgb_grid.best_estimator_.feature_importances_
lgb_importance = lgb_grid.best_estimator_.feature_importances_
# Normalize both importances for fair averaging
xgb_importance_norm = xgb_importance / xgb_importance.sum()
lgb_importance_norm = lgb_importance / lgb_importance.sum()
feature_importance = pd.DataFrame({
    'Feature': top_15_features,
    'Importance': (xgb_importance_norm + lgb_importance_norm) / 2
}).sort_values(by='Importance', ascending=False)
print("\nFinal Model Feature Importance:\n", feature_importance)
```
